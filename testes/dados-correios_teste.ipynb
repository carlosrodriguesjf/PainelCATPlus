```python
def renomear(responsavel_envio):
    lista =['Carlos Rodrigues', 'Mauro Eduardo Leopoldino', 'Vânia de paula',
       'CARLOS RODRIGUES', 'LUIZ FELIPE BRIGATTO', 'Luiz Felipe Brigatto',
       'LUIZ FELIPE', 'Luiz Felipe', 'Vânia de Paula', 'RENATA GONÇALVES',
       'Xirmane de Oliveira', 'XIRMANE DE OLIVEIRA ROCHA MARCIANO',
       'Xirmane de Oliveira Rocha Marciano.', 'Xirmane de Oliveira Rocha',
       'Xirmane de Oliveira Rocha Marciano',
       'Xirmane Oliveira Rocha Marciano', 'Xirmane Oliveira',
       'Carlos Rodriguesa', 'Carlos Rodriugues']
    if responsavel_envio in lista:
        return responsavel_envio
        

df['responsavel_envio'] = df['responsavel_envio'].apply(renomear)
df
```

```python
def correcao_nomes(responsavel_envio):
    lista_vania =['Vânia de paula','Vânia de Paula']
    lista_carlos = ['CARLOS RODRIGUES', 'Carlos Rodriguesa','Carlos Rodriugues','Carlos Rodrigues']
    lista_luiz_felipe = ['LUIZ FELIPE BRIGATTO','Luiz Felipe Brigatto','LUIZ FELIPE']
    lista_xirmane = ['XIRMANE DE OLIVEIRA ROCHA MARCIANO','Xirmane de Oliveira Rocha Marciano.','Xirmane de Oliveira Rocha','Xirmane de Oliveira Rocha Marciano','Xirmane Oliveira Rocha Marciano','Xirmane Oliveira']
    if responsavel_envio in lista_vania:
        return 'VANIA'
    elif responsavel_envio in lista_carlos:
        return 'CARLOS' 
    elif responsavel_envio == 'RENATA GONÇALVES':
        return 'RENATA'
    elif responsavel_envio in lista_xirmane:
        return 'XIRMANE'
    elif responsavel_envio in lista_luiz_felipe:
        return 'LUIZ FELIPE'
    
df['responsavel_envio'] = df['responsavel_envio'].apply(correcao_nomes)
```

```python
# Função para padronizar as datas
def padronizar_data(data):
    try:
        # Tenta converter a data para o formato padrão 'YYYY/MM/DD'
        return pd.to_datetime(data, format='%Y/%m/%d').strftime('%Y/%m/%d')
    except ValueError:
        # Em caso de erro, retorna a data original
        return data

# Aplica a função à coluna 'Data'
df['data'] = df['data'].apply(padronizar_data)
```

```python
def padronizar_data(data):
    # Tenta converter para o formato 'DD/MM/YYYY'
    try:
        return pd.to_datetime(data, format='%d/%m/%Y').strftime('%d/%m/%Y')
    except ValueError:
        pass
    
    # Tenta converter para o formato 'YYYY/MM/DD'
    try:
        return pd.to_datetime(data, format='%Y/%m/%d').strftime('%d/%m/%Y')
    except ValueError:
        pass

    # Se não for possível converter, retorna a data original
    return data

# Aplica a função à coluna 'Data'
df['data_protocolo'] = df['data'].apply(padronizar_data)
```

```python
# Convertendo a coluna 'Data' para o formato datetime
df['data_protocolo'] = pd.to_datetime(df['data_protocolo'], format='%d/%m/%Y', errors='coerce')

# Filtrando as linhas onde a data é menor ou igual a 06/06/2022
df_filtrado = df[df['data_protocolo'] >= '06-06-2022']

# Exibindo o DataFrame resultante
df_filtrado
```

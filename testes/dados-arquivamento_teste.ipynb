```python
def remove_sujeira(data_arquivamento):
    if data_arquivamento == '-------' or data_arquivamento == '----' or data_arquivamento == 'RETIRADO':
        return None
    elif data_arquivamento == '14-07-23':
         data_arquivamento = '14-07-2023'
         
         return data_arquivamento
    
    return data_arquivamento
    
    
df.data_arquivamento = df.data_arquivamento.apply(remove_sujeira)
df
```

```python
df['data_arquivamento'].unique()
```

```python
def corrige_data(data_arquivamento):
    posicao = data_arquivamento.find('-')

    if len(data_arquivamento) == 10:

        if posicao == 4:
            ano = data_arquivamento[:4]
            mes = data_arquivamento[5:7]
            dia = data_arquivamento[8:]

            return f'{dia}-{mes}-{ano}'
        

    elif len(data_arquivamento) == 8:

        if posicao == 2:
            dia = data_arquivamento[:2]
            mes = data_arquivamento[3:5]
            ano = data_arquivamento[6:]
            
            return f'{dia}-{mes}-{ano}'  
    
    return data_arquivamento
        

df.data_arquivamento = df.data_arquivamento.apply(corrige_data)

df
```

```python
import pandas as pd

def tratamento_dados(excel_file_path):


    # Use a função read_excel para ler o arquivo Excel
    xls = pd.ExcelFile(excel_file_path)

    # Obtenha a lista de nomes de sheet_names no arquivo
    sheet_names = xls.sheet_names

    # Lista de planilhas a serem ignoradas
    planilhas_a_ignorar = ['PAINEL', 'Cancelamento de matrícula','Auxiliar','PAINEL v1','PAINEL v2','PAINEL V4']

    # Itere pelas sheet_names e carregue cada uma em um DataFrame do pandas
    data = {}  # Dicionário para armazenar DataFrames de cada sheet_name

    try:
        for sheet_name in sheet_names:
            if sheet_name not in planilhas_a_ignorar:
                # Carregar o DataFrame sem cabeçalho
                nome_dataframe = xls.parse(sheet_name, header=None, skiprows=1)

                # Transforma a segunda linha em cabeçalho
                nome_dataframe.columns = nome_dataframe.iloc[0]

                # Atualiza o DataFrame excluindo a primeira linha
                nome_dataframe = nome_dataframe.iloc[1:].reset_index(drop=True)

                nome_dataframe = nome_dataframe[['ATRIBUÍDO PARA', 'STATUS', 'SETOR', 'DATA PROTOCOLO']]

                nome_dataframe.rename(columns={'ATRIBUÍDO PARA': 'nome_operador', 'STATUS': 'status', 'SETOR': 'setor', 'DATA PROTOCOLO': 'data_protocolo'}, inplace=True)

                nome_dataframe['data_protocolo'] = nome_dataframe['data_protocolo'].astype(str)

                nome_dataframe['data_protocolo'] = nome_dataframe['data_protocolo'].apply(lambda x: x.split(' ')[0])

                nome_dataframe.dropna(inplace=True)

                # Armazena o DataFrame no dicionário
                data[sheet_name] = nome_dataframe



            elif sheet_name == 'Cancelamento de matrícula':
                # Carregar o DataFrame sem cabeçalho
                nome_dataframe = xls.parse(sheet_name, header=None, skiprows=1)

                # Transforma a segunda linha em cabeçalho
                nome_dataframe.columns = nome_dataframe.iloc[0]

                # Atualiza o DataFrame excluindo a primeira linha
                nome_dataframe = nome_dataframe.iloc[1:].reset_index(drop=True)

                nome_dataframe = nome_dataframe[['ATRIBUÍDO PARA', 'STATUS', 'SETOR', 'DATA PROTOCOLO','Motivo para cancelamento','Informe o motivo, caso seja preenchido Outros no campo anterior','CAMPUS']]

                nome_dataframe.rename(columns={'ATRIBUÍDO PARA': 'nome_operador', 'STATUS': 'status', 'Motivo para cancelamento':'motivo', 'Informe o motivo, caso seja preenchido Outros no campo anterior': 'descricao_motivo','SETOR': 'setor', 'DATA PROTOCOLO': 'data_protocolo'}, inplace=True)

                nome_dataframe['data_protocolo'] = nome_dataframe['data_protocolo'].astype(str)

                nome_dataframe['data_protocolo'] = nome_dataframe['data_protocolo'].apply(lambda x: x.split(' ')[0])

                nome_dataframe.dropna(inplace=True)

                # Armazena o DataFrame no dicionário
                data[sheet_name] = nome_dataframe
    except:
        print(sheet_name)

    return data

# Especifique o caminho para o arquivo Excel
excel_file_path = r'C:\projetos\PainelCATPlus\arquivos_apoio\Distribuição de Demandas - CAT.xlsx'
dicionario = tratamento_dados(excel_file_path)

excel_file_path = r'C:\projetos\PainelCATPlus\arquivos_apoio\Distribuição de Demandas - Out2023.xlsx'
dicionario_atual = tratamento_dados(excel_file_path)


# Nome para cada item do dicionário
nomes = {'df_aprov_estudos': 'Aprov. de estudos', 'df_atestado': 'Atestado, Declaração ou Certidã', 'df_cancelamento': 'Cancelamento de matrícula', 'df_correcao': 'Correção ou Alter. de Dados Cad','df_data_especial': 'Data Especial de Col. de Grau','df_ementas': 'Ementas','df_historico': 'Histórico Escolar','df_segunda_via_graduacao': '2ª Via Diploma de Grad.','df_segubda_via_pos_graducao': '2ª Via DiplomaCertif. de Pós-Gr','df_trancamento': 'Trancamento','df_reintegracao': 'Reintegração','df_nome_social': 'Nome Social','df_restituicao': 'Restituição de taxa','df_transferencia_externa': 'Transferência(ex-officio)','df_dilatacao': 'Dilatação de curso','df_outros': 'Outros','df_nova_abi': 'Nova Graduação na mesma ABI','df_refugiados':'Ingresso de Refugiados','df_urgencia_diploma':'Urgência diploma digital'}

# Novo dicionário com nomes
dicionario_com_nomes = {}

# Itera sobre os nomes
for nome, chave_original in nomes.items():
    # Adiciona ao novo dicionário
    dicionario_com_nomes[nome] = dicionario[chave_original]


# Novo dicionário com nomes
dicionario_com_nomes_atual = {}

# Itera sobre os nomes
for nome, chave_original in nomes.items():
    # Adiciona ao novo dicionário
    dicionario_com_nomes_atual[nome] = dicionario_atual[chave_original]

nomes_dataframes = list(dicionario_com_nomes.keys())
nomes_dataframes_atual = list(dicionario_com_nomes_atual.keys())



# Itera sobre as chaves e valores do dicionário
for chave, valores in dicionario_com_nomes.items():
    # Cria um DataFrame com a chave e os valores
    df = pd.DataFrame({chave: [valores]})
    
    # Salva o DataFrame em um arquivo CSV
    df.to_csv(f'C:\\projetos\\PainelCATPlus\\dados_tratados\\antigo\\{chave}.csv', index=False)
 
# Itera sobre as chaves e valores do dicionário
for chave, valores in dicionario_com_nomes_atual.items():
    # Cria um DataFrame com a chave e os valores
    df = pd.DataFrame({chave: [valores]})
    
    # Salva o DataFrame em um arquivo CSV
    df.to_csv(f'C:\\projetos\PainelCATPlus\\dados_tratados\\atual\\{chave}.csv', index=False)
 
```

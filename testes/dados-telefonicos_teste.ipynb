```python
import pandas as pd
from datetime import date
```

```python
#carregando os arquivos
df = pd.read_excel(r'C:\projetos\PainelCATPlus\arquivos_apoio\Relatório Atendimento Telefônico (respostas).xlsx', header=0, dtype={'Carimbo de data/hora': str})
df.info()
df.head()
```

```python
df.drop(columns=['Solicitante','Anotações sobre o atendimento:'],inplace=True)
df
```

```python
df = df.dropna()
df
```

```python
df[['data','hora']] = df['Carimbo de data/hora'].str.split(' ', expand = True)
df
```

```python
# tratando datas
def tratamento_data(data):

    ano = data.split('-')[0]
    mes = data.split('-')[1]
    dia = data.split('-')[2]
    data = f'{dia}/{mes}/{ano}'

    return data

df['data'] = df['data'].apply(tratamento_data)
df
```

```python
def tratamento_hora(hora):

    nova_hora = hora[:2]
    
    return nova_hora

df['hora'] = df['hora'].apply(tratamento_hora)
df
```

```python
df.drop(columns=['Carimbo de data/hora'], inplace=True)
df.rename(columns={'Endereço de e-mail':'email','Classificação do Atendimento':'classificacao_atendimento'},inplace=True)
df
```

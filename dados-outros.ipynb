```python
#carregando os arquivos
df = pd.read_excel(r'C:\projetos\PainelCATPlus\arquivos_apoio\dados_eletronicos.xlsx', header=1)
df.head()
```

```python
df.columns[3]
```

```python
# excluindo colunas e alterando o nome da coluna
df.drop(columns=['Unnamed: 0','Unnamed: 8','Total zap','Total Jivochat'],inplace=True)


df.head()

```

```python
# tratando datas
def tratamento_data(data):

    nova_data = data.strftime('%d/%m/%y')

    return nova_data

df.insert(1, column='data', value =df['Mês'].apply(tratamento_data)) 
df.drop(columns=['Mês'], inplace=True)
```

```python
# re4nomeando colunas
novas_nomes = ['ano', 'data', 'email', 'whatsapp', 'jivochat', 'visualizacoes_site','usuarios_externos']
df.columns = novas_nomes
df
```

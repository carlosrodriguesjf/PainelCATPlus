```python
import pandas as pd
```

```python
#carregando os arquivos
df = pd.read_excel(r'C:\projetos\PainelCATPlus\arquivos_apoio\dados_funcionarios.xlsx', header=2, dtype=str)
```

```python
#excluindo excluindo coluna vazia
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
# criar a coluna setor
df['setor'] = 0
```

```python
# incluindo o setor

def define_setor(cpf):
    cpf = str(cpf)
    if cpf in ['09669452635', '08117227673','42503297803','88063631668','10070824614','08105274647']:
        return 'CDARA'
    elif cpf in ['10993501664', '07967789652','04295422673','09740750605','06422744609','11231702664','08021165669','09092059650','12150014602']:
        return 'UNIMED'
    else:
        return 'CAT'
    

df['setor'] = df['CPF'].apply(define_setor)
```

```python
# mudando o nome das colunas
df.rename(columns={'Nome':'nome','CPF':'cpf','E-mail':'email','Nome fictício':'nome_ficticio'},inplace=True)
```

```python
# excluindo os nomes (CDARA) e (U)

def limpa_nome(nome):

    if nome == 'ANA CRISTINA':
        return 'ANA CRISTINA'
    else:
        primeiro_nome = nome.split(' ') 
        return primeiro_nome[0]


df['nome'] = df['nome'].apply(limpa_nome)
```

```python
# corrigindo nome do Marcus
def corrige_nome(nome):

    if nome == 'MARCUS':
        return 'MARCUS GABRIEL'
    else:
        return nome
    
df['nome'] = df['nome'].apply(corrige_nome)
```

```python
# corrigindo o e-mail

def substituir_email(email):
    return email.replace('U.com.br', 'unimed.br')

df['email'] = df['email'].apply(substituir_email)
```

```python
# alterando a ordem das colunas

nova_ordem = ['nome', 'cpf', 'email', 'setor', 'nome_ficticio']

df = df[nova_ordem]
df
```
